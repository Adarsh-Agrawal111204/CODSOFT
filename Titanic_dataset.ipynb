```python

ports = pd.get_dummies(titanic_df.Embarked, prefix='Embarked', dtype=int)
ports.head()
```

```python
titanic_df = titanic_df.join(ports)
```

```python
titanic_df.drop(['Embarked'],axis=1, inplace=True)
     
```

```python
titanic_df.head()
```

```python
titanic_df.Sex=titanic_df.Sex.map({'male':0,'female':1})
```

```python
Y= titanic_df.Survived.copy()
X= titanic_df.drop(['Survived'], axis = 1)
```

```python
X.drop(['Cabin','Ticket', 'Name', 'PassengerId'], axis = 1, inplace=True)
```

```python
X.isnull().sum()
```

```python
Y.isnull().sum()
```

```python
X_age=np.random.randint(X['Age'].mean()-X['Age'].std(), X['Age'].mean()+X['Age'].std(), 177)
```

```python
X.loc[X['Age'].isnull(), 'Age'] = X_age
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
Y_pred_test= model.predict(X_test)
Y_pred_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred_test)
```
